```python
# Importing Libraries:
import pandas as pd
import numpy as np 
```

```python
# Reading the data
data = pd.read_csv('updated.csv', encoding= 'unicode_escape') 
data=data.drop(['Unnamed: 0'], axis=1)
data
```

Data Cleaning

```python
# Number of rows and columns in the data set
rows,cols=data.shape
```

```python
#Total number of duplicated rows
data.duplicated().sum()
```

```python
# Dropping duplicated rows
data=data.drop_duplicates()
```

```python
# Checking if there are missing data
data.isnull().values.any()
```

```python
#Missing values in the data
nullData=data.isnull().sum()
nullData
```

```python
# Percentage of missing data: ---> 3.7% of data are null values
data_count = np.product(data.shape)
(nullData.sum()/data_count)*100
```

```python
# Percentage of missing data in each column
data.isnull().sum() / data.shape[0] * 100
```

```python
# Dropping the Company column as it is 94% missing dsta
data=data.drop(['company'], axis=1)
# Dropping rows with missing data since it does not exceed 15%
data = data.dropna(axis=0)
# Ensuring that there are no missing data
data.isnull().values.any()
```

```python
#The number of nights stayed 
data['duration_of_stay'] = data['stays_in_weekend_nights'] + data['stays_in_week_nights']
```

```python
# Ensuring that there are no missing data
print(data.isnull().values.any())
# Saving the cleaned data to a csv to be used late
data.to_csv('cleaned.csv')
```

```python
# Filtering the data to get the each hotel at separate dataframe
JW_Marriott= data.copy()
x=JW_Marriott.loc[(JW_Marriott['hotel'] != 'JW Marriott Hotel')].index
#Dropping all the rows which is neither of the airlines
JW_Marriott.drop(x , inplace = True )
JW_Marriott
# Saving the JW_Marriott data as csv
JW_Marriott.to_csv('JW_Marriott.csv')
```

```python
# Filtering the data to get the each hotel at separate dataframe
Renaissance_Hotel= data.copy()
x=Renaissance_Hotel.loc[(Renaissance_Hotel['hotel'] != 'Renaissance Hotel')].index
#Dropping all the rows which is neither of the airlines
Renaissance_Hotel.drop(x , inplace = True )
Renaissance_Hotel
# Saving the Renaissance_Hotel data as csv
Renaissance_Hotel.to_csv('Renaissance_Hotel.csv')
```

```python
# Ensuring that there is no null values in the dataframes after splitting them 
print(Renaissance_Hotel.isnull().values.any())
print(JW_Marriott.isnull().values.any())
```
